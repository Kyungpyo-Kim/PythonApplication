```python
import pybithumb

## ticker
tickers = pybithumb.get_tickers()
print(tickers)
print(len(tickers))

## price
price = pybithumb.get_current_price("BTC")
print(price)

## detail
detail = pybithumb.get_market_detail("BTC")
print(detail)
desc = ["시가", "고가", "저가", "종가", "거래량"] # description
[print("00시 기준 {}: {}".format(desc[i], value)) 
    for i, value in enumerate(detail)
]

## oderbook
orderbook = pybithumb.get_orderbook("BTC")
print(orderbook)
print(orderbook['payment_currency'])
print(orderbook['order_currency'])

import datetime
ms = int(orderbook["timestamp"])
dt = datetime.datetime.fromtimestamp(ms/1000)
print(dt)

bids = orderbook['bids']
for bid in bids:
    print("매수호가: {}, 매수잔량: {}".format(bid['price'], bid['quantity'])) 

asks = orderbook['asks']
for ask in asks:
    print("매도호가: {}, 매도잔량: {}".format(ask['price'], ask['quantity']))
    
all = pybithumb.get_current_price("ALL")
for ticker, data in all.items():
    print(ticker, data['closing_price'])
```

```python
import pybithumb

btc = pybithumb.get_ohlcv("BTC")

# moving average
close = btc['close']
ma5 = close.rolling(5).mean()
last_ma5 = ma5[-2]
price = pybithumb.get_current_price("BTC")

if price > last_ma5:
    print("상승장")
else:
    print("하락장")

print(ma5)

def detectBullMarket(ticker, window=5):
    df = pybithumb.get_ohlcv(ticker)
    ma = df['close'].rolling(window=window).mean()
    cur_price = pybithumb.get_current_price(ticker)
    last_ma = ma[-2]
    if price > last_ma:
        return True
    else:
        return False

tickers = pybithumb.get_tickers()
for ticker in tickers:
    is_bull = detectBullMarket(ticker, 5)
    print("{}: {}".format(ticker, "상승장" if is_bull else "하락장"))
```
